```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ta
```

```python
df = pd.read_csv("../DATA/4-hours Pepperstone/4-hours_EURUSD.csv")
```

```python
def kama(df, col, n):
    df_copy = df.copy()
    df_copy[f"kama_{n}"] = ta.momentum.KAMAIndicator(df_copy["close"], n).kama()
    return df_copy
```

```python
df = kama(df, "close", 10)
df = kama(df, "close", 100)
```
