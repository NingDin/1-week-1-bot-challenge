```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

from tqdm import tqdm
def calculate_sadf(df, min_sample=30):
    # Vérifier que le DataFrame contient une colonne 'close'
    if 'close' not in df.columns:
        return "La colonne 'close' doit être dans le DataFrame"

    # Créer une série pour stocker les statistiques SADF
    sadf_series = pd.Series(index=df.index, dtype='float64')

    for t in tqdm(range(min_sample, len(df))):
        max_tstat = float('-inf')  # initialiser la valeur maximale du test statistique

        for t_0 in range(1, t - min_sample + 1):
            sample = df['close'].iloc[t_0:t+1].dropna()

            # Préparation des variables pour la régression
            y = sample.diff().dropna()
            X = sample.shift(1).dropna()
            X = sm.add_constant(X)

            # Ajustement du modèle de régression et récupération des statistiques
            model = sm.OLS(y, X).fit()
            t_stat = model.tvalues[1]  # t-stat pour le coefficient β

            # Mise à jour de max_tstat si nécessaire
            if t_stat > max_tstat:
                max_tstat = t_stat

        # Stocker max_tstat dans la série SADF
        sadf_series.iloc[t] = max_tstat

    return sadf_series

# Exemple d'utilisation
# df est votre DataFrame contenant une colonne 'close' avec les prix de clôture
# df = pd.read_csv('your_data.csv')
# sadf_values = calculate_sadf(df)
```

```python
df = pd.read_csv("../../../Alpha-Quant-Course/Data/FixTimeBars/AUDUSD_4H_Admiral_READY.csv")
```

```python
dft = calculate_sadf(df.iloc[0:500,:], min_sample=30)
```
