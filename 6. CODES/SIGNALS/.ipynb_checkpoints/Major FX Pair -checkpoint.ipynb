# CREATE THE DATASET
Obviously, we need to create the database for the 5 major currency pairs: <br>
EURUSD, GBPUSD, CADUSD, USDJPY, USDCHF

```python
# Import the data
eurusd = pd.read_csv("../DATA/4-hour Pepperstone/4-hour_EURUSD.csv", parse_dates=True, index_col="time").loc["2015":"2018"]
gbpusd = pd.read_csv("../DATA/4-hour Pepperstone/4-hour_GBPUSD.csv", parse_dates=True, index_col="time").loc["2015":"2018"]
usdcad = pd.read_csv("../DATA/4-hour Pepperstone/4-hour_USDCAD.csv", parse_dates=True, index_col="time").loc["2015":"2018"]
usdjpy = pd.read_csv("../DATA/4-hour Pepperstone/4-hour_USDJPY.csv", parse_dates=True, index_col="time").loc["2015":"2018"]
usdchf = pd.read_csv("../DATA/4-hour Pepperstone/4-hour_USDCHF.csv", parse_dates=True, index_col="time").loc["2015":"2018"]
```

```python
# Create the labels (that will allow us to find the interesting features that are related to them)
eurusd["time"] = eurusd.index
eurusd["time"] = eurusd["time"].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
eurusd = get_barrier(eurusd, tp=.0075, sl=-.0039, buy=True)

gbpusd["time"] = gbpusd.index
gbpusd["time"] = gbpusd["time"].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
gbpusd = get_barrier(gbpusd, tp=.0075, sl=-.0039, buy=True)

usdcad["time"] = usdcad.index
usdcad["time"] = usdcad["time"].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
usdcad = get_barrier(usdcad, tp=.0075, sl=-.0039, buy=True)

usdjpy["time"] = usdjpy.index
usdjpy["time"] = usdjpy["time"].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
usdjpy = get_barrier(usdjpy, tp=.0075, sl=-.0039, buy=True)

usdchf["time"] = usdchf.index
usdchf["time"] = usdchf["time"].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
usdchf = get_barrier(usdchf, tp=.0075, sl=-.0039, buy=True)
```

```python
eurusd["labeling"].plot()
```

```python
def create_features(df):
    df = DC_market_regime(df, 0.02)
    df = spread(df)
    df = kama_market_regime(df, "close", 60, 200)
    df = gap_detection(df)
    for i in [1,2,5,10,20,50]:
        df = auto_corr(df, "close", n=50, lag=i)
        df = log_transform(df, "close", i)
    df = derivatives(df,"close")
    df = displacement_detection(df)
    df = candle_information(df)
    df = moving_yang_zhang_estimator(df, 200)
    df = moving_parkinson_estimator(df)
    df = rolling_adf(df, "close", window_size=30)
    
    return df

eurusd = create_features(eurusd)
gbpusd = create_features(gbpusd)
usdcad = create_features(usdcad)
usdjpy = create_features(usdjpy)
usdchf = create_features(usdchf)
```

# INFORMATIONS

##### Correlation

```python
list_X = ["velocity", "acceleration", "candle_way", "filling", "spread", "tick_volume",
          'ret_log_1', 'ret_log_2', 'ret_log_5', 'ret_log_10',
       'ret_log_20', 'ret_log_50', 'autocorr_1', 'autocorr_2', 'autocorr_5',
       'autocorr_20',"autocorr_10", "kama_diff",
          "amplitude","rolling_volatility_yang_zhang", "rolling_volatility_parkinson",
         "market_regime", "kama_trend", "rolling_adf_stat"]
col_y = "labeling"
```

```python
# Make a copy of the dataframe and drop rows with NA values
correlations = []
assets = [eurusd, gbpusd, usdjpy, usdcad, usdchf]
for asset in assets:
    df_copy = asset[list_X + [col_y]].dropna()

    # Create feature matrix X and target vector y
    X = df_copy[list_X].iloc[:-1,:]
    y = df_copy[col_y].iloc[1:]

    dfc = pd.concat((X,y), axis=1)

    correlations.append(dfc.corr()["labeling"])
```

```python
correlation = pd.concat(correlations, axis=1)
correlation.columns = ["EURUSD", "GBPUSD", "USDJPY", "USDCAD", "USDCHF"]
correlation = correlation.transpose()
del correlation["labeling"]
correlation = correlation.transpose()
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='viridis', linewidths=.5)
plt.show()
```
