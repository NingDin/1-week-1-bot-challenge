```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivatives(df,col):
    """
    Docstring
    """
    
    df_copy = df.copy()

    df_copy["velocity"] = df_copy[col].diff().fillna(0)
    df_copy["acceleration"] = df_copy["velocity"].diff().fillna(0)
    
    return df_copy
```

```python
np.random.seed(0)
df = pd.DataFrame(pd.Series(np.random.randn(100).cumsum() + 100), columns=["close"])
df = derivatives(df, "close")
```

```python
plt.plot(df["velocity"])
plt.plot(df["acceleration"])
```
